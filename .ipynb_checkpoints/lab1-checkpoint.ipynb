```python
import numpy as np
import sympy as sp
import mpmath as mp
import math as mt
import time 
from matplotlib import pyplot as plt
```

```python
def getDots(solution : list, xs : list) -> list:
    fun = lambda i, x : (x ** i) * (1 - x ** 2)
    ans = []
    mx = 0
    for i in xs:
        sm = 0
        for j in range(len(solution)):
            sm += (solution[j] * fun(j,i))
        ans.append([i,sm])
        mx = max(sm,mx)
    return (ans,mx)

def collocation_method(n : int, values : list, func1 = 1, func2 = 1):
    x = sp.Symbol('x')
    i = sp.Symbol('i')
    tmp = x ** i * (1 - x ** 2)
    exp = sp.lambdify((x,i), func1 * sp.diff(tmp,x,x) + (1 + func2 * x ** 2) * tmp)
    matrix = np.zeros(shape = (n,n))
    b = [-1] * n
    for i in range(n):
        for j in range(n):
            matrix[i][j] = exp(values[i],j)
    return np.linalg.solve(matrix,b)

def galerkin_method(n : int, func1 = 1, func2 = 1):
    x = sp.Symbol('x')
    i1 = sp.Symbol('i1')
    i2 = sp.Symbol('i2')
    i3 = sp.Symbol('i3')
    tmp = x ** i1 * (1 - x ** 2)
    base = x ** i2 * (1 - x ** 2)
    cur = x ** i3 * (1 - x ** 2)
    tmp_exp = func1 * sp.diff(tmp,x,x) + (1 + func2 * x ** 2) * tmp
    aexp = sp.lambdify((x,i1,i2), tmp_exp * base)
    bexp = sp.lambdify((x,i3), cur * -1)
    points = np.linspace(-1,1,100)
    matrix = np.zeros(shape = (n,n))
    b = [0] * n
    for i in range(n):
        for j in range(n):
            matrix[i][j] = np.trapz(np.array([aexp(t,j,i) for t in points]))
        b[i] = np.trapz(np.array([bexp(t,i) for t in points]))
    return np.linalg.solve(matrix,b)

def min_squares_method(n : int, func1 = 1, func2 = 1):
    x = sp.Symbol('x')
    i = sp.Symbol('i')
    tmp = x ** i * (1 - x ** 2)
    cur = func1 * sp.diff(tmp,x,x) + (1 + func2 * x ** 2) * tmp
    exp = sp.lambdify((x,i), cur)
    
    matrix = np.zeros(shape = (n,n))
    b = [0] * n
    points = np.linspace(-1,1,n)
    
    for i in range(n):
        for j in range(n):
            matrix[i][j] = sum(exp(k,i) * exp(k,j) for k in points)
        b[i] = -sum(exp(k,i) for k in points)
    return np.linalg.solve(matrix,b)
    
def draw(res, counter : int, a : int, b : int):
    points = np.linspace(a,b,counter)
    dots, mx = getDots(res,points)
    print(mx)
    p = np.array(dots)
    x,y = p.T
    plt.plot(x,y)
    plt.show()
```

```python
st = time.time()
draw(collocation_method(100,np.linspace(-1,1,100), np.sin(8), np.cos(8)), 100, -1, 1)
end = time.time()
print(end - st)
```

```python
st = time.time()
draw(galerkin_method(100, np.sin(8), np.cos(8)),100,-1,1)
end = time.time()
print(end - st)
```

```python
st = time.time()
draw(min_squares_method(100,np.sin(8), np.cos(8)), 100, -1, 1)
end = time.time()
print(end - st)
```
